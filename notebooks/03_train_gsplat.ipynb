# 03 Train Gsplat

Run reconstruction if needed, then launch the configured gsplat command.

```python
from pathlib import Path
import sys

repo_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
src_path = repo_root / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

from pipeline.reconstruct import build_run_paths, load_reconstruction_job, run_reconstruction
from pipeline.train_gsplat import run_gsplat_training
```

```python
CONFIG_PATH = repo_root / "configs" / "house_global.yaml"
job = load_reconstruction_job(CONFIG_PATH)
paths = build_run_paths(job.config)
reconstruction = run_reconstruction(job)

if not job.config.gsplat.enabled:
    raise RuntimeError("gsplat.enabled is false in the config. Set gsplat.command before running this notebook.")

gsplat_result = run_gsplat_training(
    job.config.gsplat,
    paths=paths,
    sparse_model_path=reconstruction.sparse_model_path,
    sparse_text_path=reconstruction.sparse_text_path,
    dry_run=job.config.dry_run,
)
gsplat_result
```
